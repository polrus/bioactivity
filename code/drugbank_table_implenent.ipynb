```python
import pandas as pd
```

```python
# Creating DataFrames
bioactivity = pd.read_csv("files/med_notmoa_ge_v3.csv/med_notmoa_ge_v3.csv")
drugbank_moa = pd.read_csv("files/drugbank_click_v1.csv")
drugbank_moa
```

```python
# Group by 'drugbank_id' and concatenate values in drugbank_moa
drugbank_moa_grouped = drugbank_moa.groupby('drugbank_id').agg(lambda x: ', '.join(x))

# Reset index to make 'drugbank_id' a column again
drugbank_moa_index = drugbank_moa_grouped.reset_index()
drugbank_moa_index
```

```python
bioactivity
```

```python
# Merge the two DataFrames on 'drugbank_id'
merged_df = pd.merge(bioactivity, drugbank_moa, on='drugbank_id', how='left')

merged_df
```

```python
# Function to create the new column
def create_pharm_action_drugbank(df):
    # Create a mask where uniprot_id equals accession within each drugId
    mask = df.groupby('drugId').apply(lambda x: x['uniprot_id'] == x['accession']).reset_index(level=0, drop=True)

    # Assign pharm_action to pharm_action_drugbank where the condition is met
    df['pharm_action_drugbank'] = None
    df.loc[mask, 'pharm_action_drugbank'] = df['pharm_action']

    return df

# Apply the function to the DataFrame
df = create_pharm_action_drugbank(merged_df)
df
```

```python
bioactivity
```

```python
final_df = pd.merge(bioactivity, df[['drugId', 'uniprot_id', 'pharm_action_drugbank']], left_on=['drugId', 'accession'], right_on=['drugId', 'uniprot_id'], how='left')

final_df
```
